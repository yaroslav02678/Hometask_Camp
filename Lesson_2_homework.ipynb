```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
a = np.random.randint(1,100,20)
a
```

```python
left_number = 0
right_number = 1
lakes = []
deeps = []
```

```python
while left_number < len(a) - 1:
    if right_number < len(a) and a[right_number] < a[left_number]:
        right_number += 1 
    else:
        lake = a[left_number : right_number + 1]
        if len(lake) > 2:
            lakes.append(lake)
        left_number += 1
        right_number = left_number + 1
```

```python
for i, lake in enumerate(lakes):
    print(f"Озеро {i + 1}: {lake}")
```

```python
for i, lake in enumerate(lakes):
    if lake[0] < lake[-1]:
        bottom = min(lake[:-1])
        deep = lake[0] - bottom
        deeps.append(deep)
    else:
        bottom = min(lake[1:])
        deep = lake[-1] - bottom
        deeps.append(deep)
```

```python
for i, deep in enumerate(deeps):
    print(f"Грибина озера {i + 1}: {deep}")
```

```python
print("Найбільша глибина - ", max(deeps))
need_lake = lakes[deeps.index(max(deeps))]
print("Озеро - ", need_lake)
left_number = need_lake[0]
print("Лівий берег - ", left_number)
right_number = need_lake[-1]
print("Правий берег - ", right_number)
bottom = min(need_lake) 
print("Рівень дна - ", bottom)
if left_number > right_number:
    print("Рівень води - ", left_number)
else:
    print("Рівень води - ", right_number)
```

```python
index_left_number = np.where(a == left_number)[0][0]
index_right_number = np.where(a == right_number)[0][0] + 1
```

### Тут важливо вказати коефіціенти та область з головного графіку

```python
res = {
    'l': index_left_number,
    'r': index_right_number, 
    'lake':  a[index_left_number:index_right_number]  
}
```

```python
plt.plot (a)
plt.plot (range(res['l'],res['r']),res['lake'], c='red')
```
