```python
import pandas as pd
import numpy as np
import re
```

```python
energy = pd.read_excel('Lesson_3_hometask_data/Energy Indicators.xls', header = 17,
                                                skipfooter = 38)
```

```python
energy.drop(columns=['Unnamed: 0', 'Unnamed: 1'], inplace=True)
```

```python
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
energy['Energy Supply'] = energy['Energy Supply'] * 1_000_000
```

```python
energy.replace(to_replace=r'\.{2,}', value=np.NaN, regex=True, inplace=True)
pd.set_option('display.float_format', '{:,.0f}'.format)
```

```python
country_rename_dict = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}

energy['Country'] = energy['Country'].replace(country_rename_dict)
```

```python
energy['Country'] = energy['Country'].apply(lambda x: re.sub(r'\(.*\)', '', x))  
energy['Country'] = energy['Country'].apply(lambda x: re.sub(r'\d+', '', x))
```

```python
GDP = pd.read_csv('Lesson_3_hometask_data/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_76261.csv', header = 2)
```

```python
country_rename_dict = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}

GDP['Country Name'] = GDP['Country Name'].replace(country_rename_dict)
```

```python
GDP = GDP[['Country Name'] + [str(year) for year in range(2006, 2016)]]
ScimEn_top15 = scimEn[scimEn['Rank'] <= 15]
merged = pd.merge(energy, scimEn_top15, left_on='Country', right_on='Country', how='inner')
final_merged = pd.merge(merged, GDP, left_on='Country', right_on='Country Name', how='inner')
```

```python
final_merged.set_index('Country', inplace=True)
columns_to_keep = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
                   'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
                   '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
final_merged = final_merged[columns_to_keep]
```

```python
def answer_one():
    return final_merged
```

```python
answer_one()
```

```python
def answer_two():
    Top15 = answer_one()
    gdp_columns = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[gdp_columns].mean(axis=1, skipna=True)

    return avgGDP.sort_values(ascending=False)
```

```python
answer_two()
```

```python
def answer_three():
    Top15 = answer_one()
    gdp_columns = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[gdp_columns].mean(axis=1, skipna=True)
    country = avgGDP.sort_values(ascending=False).index[5]
    
    gdp_2006 = Top15.loc[country, '2006']
    gdp_2015 = Top15.loc[country, '2015']
    
    gdp_change = gdp_2015 - gdp_2006
    
    return gdp_change
```

```python
answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    Top15['Self-Citations to Total Citations Ratio'] = Top15['Self-citations'] / Top15['Citations']
    max_ratio = Top15['Self-Citations to Total Citations Ratio'].max()
    country_with_max_ratio = Top15['Self-Citations to Total Citations Ratio'].idxmax()
    
    return (country_with_max_ratio, max_ratio)
```

```python
answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['Estimated Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    sorted_top15 = Top15.sort_values('Estimated Population', ascending=False)
    
    return sorted_top15.index[2]
```

```python
answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['Citable Documents per Capita'] = Top15['Citable documents'] / Top15['Energy Supply per Capita']
    correlation = Top15['Citable Documents per Capita'].corr(Top15['Energy Supply per Capita'])
    
    return correlation
```

```python
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict = {'China':'Asia', 
                     'United States':'North America', 
                     'Japan':'Asia', 
                     'United Kingdom':'Europe', 
                     'Russian Federation':'Europe', 
                     'Canada':'North America', 
                     'Germany':'Europe', 
                     'India':'Asia',
                     'France':'Europe', 
                     'South Korea':'Asia', 
                     'Italy':'Europe', 
                     'Spain':'Europe', 
                     'Iran':'Asia',
                     'Australia':'Australia', 
                     'Brazil':'South America'}
    Top15['Continent'] = Top15.index.map(ContinentDict)
    Top15['Estimated Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    continent_stats = Top15.groupby('Continent')['Estimated Population'].agg(
        size='size',
        sum='sum',
        mean='mean',
        std='std'
    )
    
    return continent_stats
```
